
Convert price_.df, volume_.df into one single dataset


```python
def import_df(path, crop, method):
    price_df = pd.read_csv(f"{path}/price_{crop}_{method}.csv", index_col=0, parse_dates=True, low_memory=False)
    volume_df = pd.read_csv(f"{path}/volume_{crop}_{method}.csv", index_col=0, parse_dates=True, low_memory=False)
    return price_df, volume_df
```

```python
crop = 'brinjal'
path = './'
# import dataframe
price_df, volume_df = import_df(path, crop, 'locf')
```

```python
# init new df
crop_df = pd.DataFrame(columns={'Date', 'Market', 'Crop', 'Volume', 'Price'}, index=range(0))

# markets list 
markets = price_df.columns
```

```python
i = 0
for market in markets:
    df = pd.DataFrame({'Date': pd.date_range("2008-1-1", "2018-12-31"), 'Market': market, 'Crop': crop, 
                       'Volume': volume_df[market], 'Price': price_df[market]})
    crop_df = pd.concat([df, crop_df], ignore_index=True)
#     i+=1
#     if i == 5:
#         break
print(crop_df)

```
